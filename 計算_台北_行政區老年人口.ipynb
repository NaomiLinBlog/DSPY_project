```python
import pandas as pd
# 載入數據
file_path = 'data/112年1-12月各區按年齡人口數.csv'  # 替換為您的檔案路徑
data = pd.read_csv(file_path)
column_names = data.columns

# 找出所有含有 "合計" 和 '100歲以上' 的欄位名称
columns_to_keep = [col for col in column_names if '合計' in col or '100歲以上' in col]
columns_to_keep = ['區 域 別'] + columns_to_keep  # 加入區域別欄位
filtered_data=data[columns_to_keep]
# 刪除性別為 '男' 或 '女' 的行
filtered_data = filtered_data[(data['性別'] != '男') & (data['性別'] != '女')]
filtered_data.head()
```

```python
# 刪除所有索引小於 '合計_65~69歲' 的欄位
age_65_sum_index = filtered_data.columns.get_loc("合計_65~69歲")
columns_to_keep_after_65_sum = ['區 域 別'] + list(filtered_data.columns[age_65_sum_index:])

filtered_data_final = filtered_data[columns_to_keep_after_65_sum]
# 新增合計欄位，計算65歲以上的總人口
filtered_data_final['65歲以上總和'] = filtered_data_final.iloc[:, 1:].sum(axis=1)

filtered_data_final = filtered_data_final[filtered_data_final['區 域 別'] != '總  計']
filtered_data_final = filtered_data_final.rename(columns={'區 域 別': '區域別'})
# 顯示結果
filtered_data_final.head()
```

```python

filtered_data_final.to_csv('data/台北_行政區老年人口.csv')
```
